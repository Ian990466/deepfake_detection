# EfficientNet Visualiztion

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import roc_curve, roc_auc_score, auc
```

# ROC_AUC Curve 

```python
df = pd.read_pickle("../results/net-EfficientNetBA_traindb-dfdc-35-5-10_face-scale_size-224_seed-41_bestval/dfdc-35-5-10_test.pkl")

# print(df.columns)

# d_label = df['label'].astype(int).to_numpy()
# d_score = df["score"].to_numpy()
# d_value = df["loss"].to_numpy()
# d_label
```

```python
fpr, tpr, threshold = roc_curve(df['label'].astype("int"), df['score'])
print(fpr, tpr, threshold)
```

```python
roc_auc = auc(fpr, tpr)
## Plot the result
plt.figure(figsize=(16,9))
plt.title('Testset ROC AUC Curve')
plt.plot(fpr, tpr, color = 'orange', label = "ROC curve(area = %0.2f)" % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()    
```
